```
import requests
from bs4 import BeautifulSoup
from time import sleep

page = requests.get('https://www.letras.mus.br/mais-acessadas/axe/')

# Criar o objeto BeautifulSoup
soup = BeautifulSoup(page.text, 'html.parser')
```

```
music_tag_list = soup.find_all('li')
```

```
music_list = []
for music_tag in music_tag_list:
  if 'title' in music_tag.contents[0].attrs:
    artist = music_tag.contents[0].contents[3].string
    music = music_tag.contents[0]['title']
    href = music_tag.contents[0]['href']
    music_list.append((artist, music, 'https://www.letras.mus.br' + href))
  else:
    print('erro:', music_tag)
```

```
music_list
```

```
sleep_step_time = 0.5
error_sleep_time = 10
lyrics_list = []

for ix, m in enumerate(music_list):
  if ix % 50 == 0:
    print('page', ix, 'of', len(music_list))
  try:
    page = requests.get(m[2])
  except:
    print('error_page_ix:', ix)
    print('error_page:', m)
    sleep(error_sleep_time)
    sleep_step_time *= 2

  soup = BeautifulSoup(page.text, 'html.parser')
  lyric_tag = soup.find_all('div', class_='cnt-letra p402_premium')
  if lyric_tag == []:
    print
    continue
  lyric = ''.join([str(c) for c in lyric_tag[0].contents]).strip()
  lyrics_list.append((m[0], m[1], m[2], lyric))
  sleep(sleep_step_time)
```
